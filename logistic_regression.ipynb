```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("C:\\Users\\Hp\\Desktop\\forest fire detection\\Forest_fire.csv")
data= dataset.copy()
data
```

```python
data.drop(columns='Area')
```

```python
O=pd.array(data["Oxygen"])
T=pd.array(data["Temperature"])
H=pd.array(data["Humidity"])
F=pd.array(data["Fire Occurrence"])


```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(8, 6))
axes[0].scatter(O,F, label='oxygen')
axes[0].set_title('oxygen/fire')
axes[0].legend()

axes[1].scatter(T,F, label='temperature', color='orange')
axes[1].set_title('temperature/fire')
axes[1].legend()

axes[2].scatter(H,F, label='Humidity', color='red')
axes[2].set_title('Humidity/fire')
axes[2].legend()


plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
logreg=LogisticRegression()
```

```python
X=data[['Oxygen','Temperature','Humidity']]
Y=data['Fire Occurrence']
```

```python
plt.scatter(X['Temperature'], X['Oxygen'],X['Humidity'],c=Y, cmap='viridis', marker='o')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
logreg.fit(X_train,y_train)
```

```python
logreg.score(X_train,y_train)
```

```python
y_pred=logreg.predict(X_test)
y_pred
```
